```python
Adj_close_price=stockdata[['Adj Close']]
Adj_close_price
```

```python
min(Adj_close_price.values),max(Adj_close_price.values)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler(feature_range=(0,1))
scaled_data= scaler.fit_transform(Adj_close_price)
scaled_data
```

```python
x_data=[]
y_data=[]

for i in range(100,len(scaled_data)):
    x_data.append(scaled_data[i-100:i])
    y_data.append(scaled_data[i])

x_data,y_data=np.array(x_data),np.array(y_data)
```

```python
x_data[0],y_data[0]
```

```python
train_size=len(x_data)*0.7
import math
train_size=math.floor(train_size)
train_size
```

```python
x_train=x_data[:train_size]
x_test=x_data[train_size:]
y_train=y_data[:train_size]
y_test=y_data[train_size:]

#Shapes
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(y_test.shape)
```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, Activation
```

```python

def create_simple_lstm_model(input_shape):
    model = Sequential()
    model.add(LSTM(64, input_shape=input_shape, return_sequences=True))
    model.add(LSTM(32))
    model.add(Dense(1))
    model.add(Activation('linear'))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

input_shape = (x_train.shape[1], 1)
model = create_simple_lstm_model(input_shape)
```

```python
# Compile the model
model.compile(optimizer='adam', loss='mean_squared_error')

# Fit the model
model.fit(x_train, y_train, batch_size=1, epochs=2)
```

```python
model.summary()
```

```python
predictions = model.predict(x_test)
predictions
```

```python
inv_predictions=scaler.inverse_transform(predictions)
```

```python
inv_y_test=scaler.inverse_transform(y_test)
inv_y_test
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

# Mean Absolute Error
mae = mean_absolute_error(inv_y_test, inv_predictions)
print(f"Mean Absolute Error (MAE): {mae}")

# Mean Squared Error
mse = mean_squared_error(inv_y_test, inv_predictions)
print(f"Mean Squared Error (MSE): {mse}")

# Root Mean Squared Error
rmse = mean_squared_error(inv_y_test, inv_predictions, squared=False)
print(f"Root Mean Squared Error (RMSE): {rmse}")

# Mean Absolute Percentage Error
mape = np.mean(np.abs((inv_y_test - inv_predictions) / inv_y_test)) * 100
print(f"Mean Absolute Percentage Error (MAPE): {mape}%")

# R-squared Score
r2 = r2_score(inv_y_test, inv_predictions)
print(f"R-squared (R²) Score: {r2}")
```
